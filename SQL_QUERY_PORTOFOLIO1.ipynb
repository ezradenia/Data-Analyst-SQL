#### Case 1

Pemilik toko mengalami kesulitan saat melihat tabel Orders karena informasi customer dan product hanya berupa ID, sedangkan pemilik toko ingin melihat nama customer beserta kode dan tipe kendaraan yang dibeli

```python
#Penggunaan Join 
pd.read_sql_query("""  
                  SELECT
                  c.CUSTOMER_ID,
                  c.CUSTOMERNAME,
                  o.*,
                  p.PRODUCTCODE,
                  p.PRODUCTLINE
                  FROM orders o
                  LEFT JOIN Products p ON o.PRODUCT_ID = p.PRODUCT_ID
                  LEFT JOIN customers c ON o.CUSTOMER_ID = c.CUSTOMER_ID""",conn)
```

#### Case 2

Pemilik toko ingin mengetahui apakah terjadi peningkatan atau bahkan penurunan baik jumlah customer maupun jumlah transaksi pembelian di setiap tahunnya, selain itu ingin mengetahui apakah terjadi New Purchase (Pembelian Baru) dari New Customer. Untuk mengetahui hal tersebut, pertama-tama kita harus mendapatkan berapa jumlah uniq customer di setiap tahunnya.

```python
#mendapatkan berapa jumlah uniq customer_id di setiap tahunnya
pd.read_sql_query("""  
                  SELECT
                    strftime('%Y',ORDERDATE) AS Year,
                    COUNT(DISTINCT CUSTOMER_ID) AS Customer_Size
                  FROM Orders
                  GROUP BY strftime('%Y',ORDERDATE)""",conn)
```

```python
#mendapatkan berapa jumlah uniq order_id di setiap tahunnya
pd.read_sql_query("""
                    SELECT 
                        strftime('%Y',ORDERDATE) AS Year,
                        COUNT(DISTINCT ORDER_ID) AS '# of Orders'
                    FROM Orders
                    GROUP BY 1
                  """, conn)
```

Berdasarkan dua code diatas, didapatkan bahwa terjadi penurunan baik jumlah customer yang melakukan transaksi pembelian maupun banyaknya transaksi pembelian di toko tersebut setiap tahunnya, terlebih lagi penurunan yang signifikan di tahun 2005.

Apabila terjadi New Purchase di tahun 2004 maupun 2005, pemilik toko ingin mengetahui semua informasi tentang New Customer tersebut.

```python
# Pertama, mencari tanggal pertama transaksi setiap customer
pd.read_sql_query("""
                    SELECT 
                        CUSTOMER_ID,
                        DATE(MIN(ORDERDATE)) AS DATE_OF_FIRST_TRANSC
                    FROM Orders
                    GROUP BY 1
                    ORDER BY DATE_OF_FIRST_TRANSC ASC
                  """, conn)
```

```python
# Lalu, hitung banyaknya customer yang lahir di setiap tahunnya
pd.read_sql_query("""
                    SELECT 
                        strftime('%Y',dob.DATE_OF_FIRST_TRX) AS YEAR_OF_FIRST_TRX,
                        COUNT(DISTINCT CUSTOMER_ID) AS NEW_CUSTOMER_SIZE
                    FROM(
                        SELECT 
                            CUSTOMER_ID,
                            DATE(MIN(ORDERDATE)) AS DATE_OF_FIRST_TRX
                        FROM Orders
                        GROUP BY 1) dob
                    GROUP BY 1
                  """, conn)
```

Maka diketahui tidak ada New Purchase (Pembelian baru dari Customer baru) pada tahun 2005.

```python
# Selanjutnya, didapatkan semua informasi customer yang melakukan pembelian pertama kali di tahun 2004
pd.read_sql_query("""
                    SELECT
                    c.*
                    FROM Customers c
                    LEFT JOIN
                        (SELECT 
                            CUSTOMER_ID,
                            DATE(MIN(ORDERDATE)) AS DATE_OF_FIRST_TRANSC
                        FROM Orders
                        GROUP BY 1
                        ORDER BY DATE_OF_FIRST_TRANSC ASC) dob
                        ON dob.CUSTOMER_ID = c.CUSTOMER_ID
                        GROUP BY 1,2,3,4,5,6,7,8,9
                        HAVING strftime('%Y',dob.DATE_OF_FIRST_TRANSC) = '2004'
                  """, conn)
```

#### Case 3

Sebutkan top 3 product pada tahun 2003. Apakah dari produk-produk tersebut ada yang menjadi favorit dalam 3 tahun berturut-turut?

```python
pd.read_sql_query("""  
                  SELECT *
                  FROM
                    (SELECT
                    strftime('%Y',ORDERDATE) AS Year,
                    PRODUCT_ID,
                    SUM(QUANTITYORDERED) AS TOTAL_ORDERED
                  FROM Orders
                  WHERE strftime('%Y',ORDERDATE) = '2003'
                  GROUP BY PRODUCT_ID
                  ORDER BY TOTAL_ORDERED DESC
                  LIMIT 3)

                  UNION 
                  
                  SELECT*
                  FROM
                    (SELECT
                    strftime('%Y',ORDERDATE) AS Year,
                    PRODUCT_ID,
                    SUM(QUANTITYORDERED) AS TOTAL_ORDERED
                  FROM Orders
                  WHERE strftime('%Y',ORDERDATE) = '2004'
                  GROUP BY PRODUCT_ID
                  ORDER BY TOTAL_ORDERED DESC
                  LIMIT 3)

                  UNION 
                   SELECT*
                  FROM
                    (SELECT
                    strftime('%Y',ORDERDATE) AS Year,
                    PRODUCT_ID,
                    SUM(QUANTITYORDERED) AS TOTAL_ORDERED
                  FROM Orders
                  WHERE strftime('%Y',ORDERDATE) = '2005'
                  GROUP BY PRODUCT_ID
                  ORDER BY TOTAL_ORDERED DESC
                  LIMIT 3)

                  ORDER BY 3 DESC
                  """,conn)
```

#### Case 4

Dapatkan data total penjualan dengan status order dalam proses pengiriman (shipped), ditahan (on hold), dan dibatalkan (cancelled) untuk setiap customer.

```python
#Mendapatkan informasi Total Sales dari setiap Customer yang status order nya "Shipped"
pd.read_sql_query("""  
                  SELECT 
                    CUSTOMER_ID,
                    SUM(SALES) AS 'Sales Shipped'
                  FROM Orders
                  WHERE STATUS = 'Shipped'
                  GROUP BY CUSTOMER_ID
                  """,conn)
```

```python
#Mendapatkan informasi Total Sales dari setiap Customer yang status order nya "Cancelled"
pd.read_sql_query("""  
                  SELECT 
                    CUSTOMER_ID,
                    SUM(SALES) AS 'Sales Cancelled'
                  FROM Orders
                  WHERE STATUS = 'Cancelled'
                  GROUP BY CUSTOMER_ID
                  """,conn)
```

```python
#Mendapatkan informasi Total Sales dari setiap Customer yang status order nya "On Hold"
pd.read_sql_query("""  
                  SELECT 
                    CUSTOMER_ID,
                    SUM(SALES) AS 'Sales on Hold'
                  FROM Orders
                  WHERE STATUS = 'On Hold'
                  GROUP BY CUSTOMER_ID
                  """,conn)
```

```python
#Menggabungkan 3 informasi di atas kedalam satu tabel agar lebih mudah dibaca
pd.read_sql_query("""  
                  SELECT
                   c.Customer_ID,
                   c.CUSTOMERNAME,
                   CASE WHEN ss.Sales_Shipped IS NULL THEN 0 ELSE ss.Sales_Shipped END AS 'Sales Shipped',
                   CASE WHEN cl.Sales_Cancelled IS NULL THEN 0 ELSE cl.Sales_Cancelled END AS 'Sales Cancelled',
                   CASE WHEN oh.Sales_On_Hold IS NULL THEN 0 ELSE oh.Sales_On_Hold END AS 'Sales On Hold'
                  FROM Customers c
                  LEFT JOIN
                    (SELECT 
                      CUSTOMER_ID,
                      SUM(SALES) AS Sales_Shipped
                    FROM Orders
                    WHERE STATUS = 'Shipped'
                    GROUP BY CUSTOMER_ID)ss
                  ON ss.Customer_ID = c.Customer_ID
                  LEFT JOIN
                    (SELECT 
                      CUSTOMER_ID,
                      SUM(SALES) AS Sales_Cancelled
                    FROM Orders
                    WHERE STATUS = 'Cancelled'
                    GROUP BY CUSTOMER_ID) cl
                  ON c.Customer_ID = cl.Customer_ID
                  LEFT JOIN
                    (SELECT 
                      CUSTOMER_ID,
                      SUM(SALES) AS Sales_On_Hold
                    FROM Orders
                    WHERE STATUS = 'On Hold'
                    GROUP BY CUSTOMER_ID) oh
                  ON c.Customer_ID = oh.Customer_ID
                  """,conn)
```

#### Case 5

Berapa jumlah order quantity yang terjual masing-masing produk pada tahun 2004?

```python
pd.read_sql_query(""" 
                  SELECT 
                  PRODUCT_ID,
                  SUM(QUANTITYORDERED)
                  FROM 
                  Orders
                  WHERE strftime('%Y',ORDERDATE) ='2004'
                  GROUP BY PRODUCT_ID
                 """,conn)
```

Pemilik toko ingin mengetahui hasil penjualan di setiap tahunnya. Oleh karena itu dari tabel Orders, berapakah pendapatan di setiap tahun?

```python
pd.read_sql_query(""" 
                  SELECT 
                  strftime('%Y',ORDERDATE) AS Year,
                  SUM(SALES)
                  FROM 
                  Orders
                  GROUP BY Year
                 """,conn)
```

Pada tahun 2003 Pemilik toko menjumpai anomali pada database, Pemilik toko mendapati masih terdapat order yang berstatus "On Hold", untuk menyelidiki hal tersebut pemilik toko ingin seluruh informasi order pada tahun 2003 yang berstatus "On Hold".
